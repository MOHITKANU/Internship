# top 100 indian movies

```python
from bs4 import BeautifulSoup
import requests
import re
 
url = 'https://www.imdb.com/india/top-rated-indian-movies/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
 
movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
 
ratings = [b.attrs.get('data-value')
           for b in soup.select('td.posterColumn span[name=ir]')]
 
votes = [b.attrs.get('data-value')
         for b in soup.select('td.ratingColumn strong')]
 
list = []
 
list = []
 
for index in range(0, len(movies)):
     
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "year": year,
            "place": place,
            "star_cast": crew[index],
            "rating": ratings[index],
            "vote": votes[index],
            "link": links[index]}
    list.append(data)
for movie in list:
    print(movie['movie_title'], '('+movie['year'] +
          ') -', movie['rating'])
```

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['ratings'] = ratings
data['year'] = year
data[ :101]
```

# top 10 team in cricket

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
icc=BeautifulSoup(page.content)
Team_name = []
points = []

for i in icc.find_all('span', class_="u-hide-phablet" ):
    Team_name.append(i.text)

matches=[]
for i in icc.findAll('td', class_="rankings-block__banner--matches"):
    matches.append(i.text)
for i in icc.findAll('td', class_="table-body__cell u-center-text"):
    matches.append(i.text.split())
    
rating=[]
for i in icc.findAll('td', class_="rankings-block__banner--rating u-text-right"):
    rating.append(i.text.split())
for i in icc.find_all('td', class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
       
df=pd.DataFrame[{'Team':Team_name, 'matches':matches, 'rating':rating}]
df
```

# bewakoof website data scrape program

```python
import pandas as pd
from bs4 import BeautifulSoup
Bewakoof=requests.get("https://www.bewakoof.com/women-tshirts?ga_q=tshirts")

soup = BeautifulSoup(Bewakoof.content)

product_details = []
for i in soup.find_all('div',class_="productCardDetail" ):
    product_details.append(i.text)
images = []
for i in soup.find_all('img', class_="productImgTag"):
    images.append(i['src'])
price = []
for i in soup.find_all('span', class_="discountedPriceText"):
    price.append(i.text)
df= pd.DataFrame({'product details':product_details,'price' :price,'images':images})
df
```

# dineout data scrape

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
dineout=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")

soup = BeautifulSoup(dineout.content)

dineout = []
for i in soup.find_all('div',class_= "restnt-info cursor"):
    dineout.append(i.text)

rating = []
for i in soup.find_all('div', class_="restnt-rating rating-4"):
    rating.append(i.text)

images = []
for i in soup.find_all('img', class_="no-img"):
    images.append(i['data-src'])

location = []
for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    location.append(i.text)

cuisine = []
for i in soup.find_all('div', class_="detail-info"):
    cuisine.append(i.text)

df= pd.DataFrame({'dineout':dineout,'location' :location,'cuisine' :cuisine,'rating' :rating,'images':images})
df
```
